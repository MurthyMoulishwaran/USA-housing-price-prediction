```python
from sklearn.model_selection import train_test_split as tts
```

```python
x=usa[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population']]
y=usa.Price
x
y
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.4,random_state=200)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.base import RegressorMixin as rm

lr = LR()
lr.fit(x_train,y_train)
```

```python
pre=lr.predict(x_test)
pre
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error,f1_score,mean_absolute_error,mean_squared_error
score = r2_score(y_test,pre)*100
score
```
